```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pd.read_csv("resumes.csv")

print(data)
```

```python
#you can change your job description here 
job_description = "Python Machine Learning Data Analysis Pandas"
documents = list(data["resume"]) + [job_description]
```

```python
vectorizer = TfidfVectorizer()

matrix = vectorizer.fit_transform(documents)
```

```python
similarity = cosine_similarity(matrix[-1], matrix[:-1])
```

```python
data["match_score"] = similarity[0] * 100
```

```python
data = data.sort_values(by="match_score", ascending=False)

print(data)
```

```python
job_skills = job_description.lower().split()

def missing_skills(resume):
    resume_words = resume.lower().split()
    return [skill for skill in job_skills if skill not in resume_words]

data["missing_skills"] = data["resume"].apply(missing_skills)
print(data[["name","match_score","missing_skills"]])
```
